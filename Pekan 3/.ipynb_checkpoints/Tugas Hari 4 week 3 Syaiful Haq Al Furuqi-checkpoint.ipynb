```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

<h2>Soal 1: Visualisasi Scatter Plot Perbandingan Kuantitatif</h2>

Pada tugas kali ini kita akan mengamati nilai DewPointAvg (F) dengan mengamati nilai  HumidityAvg (%), TempAvg (F), dan WindAvg (MPG)

Perhatikan bahwa data kita tidaklah siap untuk di analisis, salah satunya tipe data dari DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPG) adalah object, padahalnya data nya ber isi numeric. maka :
- Ubahlah tipe data tersebut menjadi tipe data float

Kemudian: 

- Kalian tidak akan dengan mudah mengubah tipe data tersebut karena column tersebut mempunyai nilai '-' yang dimana tidak bisa di ubah ke bentuk float, maka replace lah terlebih dahulu data yang bernilai '-' dengan nilai NaN, gunakan method .replace(). baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html
- Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
- Sekarang ubah tipe datanya dengan float, gunakan method .astype(), baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html

Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka:

Buahlah visualisasi perbandingan kuantitatif scatter plot, sehingga menghasilkan gambar seperti dibawah:

ket: 

- colormap adalah 'coolwarm'
- berikat warna terhadap setiap data poin dengan nilai dari column TempAvgF
- berikan size terhadap setiap data poin dengan nilai dari column WindAvgMPH, kalikan dengan 20 agar size terlihat lebih besar


Berikan pendapat dari insight yang bisa di dapat dari visualisasi perbandingan kuantitatif ini!!!

![alt text](https://drive.google.com/uc?id=1Pp5oIc-39mcw2vrw6xgCRM9GsvyRtPI5)

```python
# code here
df.replace({'-':np.nan}, inplace=True)
df.fillna(method='ffill')
df[['DewPointAvgF', 'HumidityAvgPercent', 'WindAvgMPH']] = df[['DewPointAvgF', 'HumidityAvgPercent', 'WindAvgMPH']].astype('float')

colors=df['TempAvgF']
size=df['WindAvgMPH']*20
fig, ax = plt.subplots(figsize=(17, 8))

cbar = ax.scatter(df['HumidityAvgPercent'], df['DewPointAvgF'], c=colors, s=size, cmap='coolwarm')
ax.set_xlabel('Humidity Avg %')
ax.set_ylabel('Dew Point Avg (F)')
ax.set_title('Austin Weather')
fig.colorbar(cbar)

plt.show()
```

Dari data tersebut dapat dilihat tentang sebaran cuaca di Austin, bahwa pada data jika semakin biru, maka cuaca disana semakin lembab, atau semakin sejuk, dan jika semakin merah, semakin kering.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 2: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
data = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
data
```

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
fig, ax = plt.subplots(figsize=(16, 8))

x = np.arange(0, 4.8, 1.2)
y1 = data.loc['Action']
y2 = data.loc['Adventure']
y3 = data.loc['Fighting']
y4 = data.loc['Misc']
y5 = data.loc['Platform']
y6 = data.loc['Puzzle']
y7 = data.loc['Racing']
y8 = data.loc['Role-Playing']
y9 = data.loc['Shooter']
y10 = data.loc['Simulation']
y11 = data.loc['Sports']
y12 = data.loc['Strategy']

ax.bar(x, y1, width=0.08, color='red', label='Action')
ax.bar(x+0.08, y2, width=0.08, color='blue', label='Adventure')
ax.bar(x+0.16, y3, width=0.08, color='purple', label='Fighting')
ax.bar(x+0.24, y4, width=0.08, color='grey', label='Misc')
ax.bar(x+0.32, y5, width=0.08, color='gold', label='Platform')
ax.bar(x+0.4, y6, width=0.08, color='green', label='Puzzle')
ax.bar(x+0.48, y7, width=0.08, color='pink', label='Racing')
ax.bar(x+0.56, y8, width=0.08, color='red', label='Role-Playing')
ax.bar(x+0.64, y9, width=0.08, color='blue', label='Shooter')
ax.bar(x+0.72, y10, width=0.08, color='purple', label='Simulation')
ax.bar(x+0.80, y11, width=0.08, color='grey', label='Sports')
ax.bar(x+0.88, y12, width=0.08, color='yellow', label='Strategy')

ax.set_title('Mean Sales Video Games By Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
labels = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
ax.set_xticks(x+0.4)
ax.set_xticklabels(labels)
ax.legend(loc='upper right', bbox_to_anchor=(0.55, 0.67, 0.5, 0.5), frameon=True)
plt.style.use('seaborn')

plt.show()
```

Jawab:<br/>
Visualisasi data tersebut tentang rata-rata penjualan video games per lokasi, pada visualisasi tersebut dapat diketahui bahwa yang paling tinggi adalah penjualan video games yang ber-<i>genre</i> <b>Platform</b> pada NA atau North America, dan yang terendah adalah <b>Shooter</b> pada selain NA, EU(Europe), JP(Japan).<br/>
Menurut saya data tersebut cukup jelas, terdapat legend juga untuk penanda warna berdasarkan <i>genre</i>.

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
fig, ax = plt.subplots(figsize=(16, 8))
x = np.arange(0, 8, 2)

ax.bar(x, y1, bottom=[0,0,0,0], color='red', width=1.5, label='Action')
ax.bar(x, y2, bottom=y1, color='blue', width=1.5, label='Adventure')
ax.bar(x, y3, bottom=y2+y1, color='purple', width=1.5, label='Fighting')
ax.bar(x, y4, bottom=y3+y2+y1, color='grey', width=1.5, label='Misc')
ax.bar(x, y5, bottom=y4+y3+y2+y1, color='gold', width=1.5, label='Platform')
ax.bar(x, y6, bottom=y5+y4+y3+y2+y1, color='green', width=1.5, label='Puzzle')
ax.bar(x, y7, bottom=y6+y5+y4+y3+y2+y1, color='pink', width=1.5, label='Racing')
ax.bar(x, y8, bottom=y7+y6+y5+y4+y3+y2+y1, color='red', width=1.5, label='Role-Playing')
ax.bar(x, y9, bottom=y8+y7+y6+y5+y4+y3+y2+y1, color='blue', width=1.5, label='Shooter')
ax.bar(x, y10, bottom=y9+y8+y7+y6+y5+y4+y3+y2+y1, color='purple', width=1.5, label='Simulation')
ax.bar(x, y11, bottom=y10+y9+y8+y7+y6+y5+y4+y3+y2+y1, color='grey', width=1.5, label='Sports')
ax.bar(x, y12, bottom=y11+y10+y9+y8+y7+y6+y5+y4+y3+y2+y1, color='yellow', width=1.5, label='Strategy')

ax.set_title('Mean Sales Video Games By Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
labels = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc='upper right', bbox_to_anchor=(0.55, 0.67, 0.5, 0.5), frameon=True)
plt.style.use('seaborn')

plt.show()
```

Jawab:<br/>
Visualisasi tersebut sama dengan visualisasi sebelumnya tentang penjualan video games pada masing-masing tempat, akan tetapi bertumpuk, dan menurut saya jika langsung disajikan pada klien, mereka mungkin cukup susah mengertinya, jadi kita harus jelaskan terlebih dahulu, atau dapat pula jika ingin lebih detail dan mudah dibaca, maka dapat menggunakan visualisasi data yang sebelumnya.
