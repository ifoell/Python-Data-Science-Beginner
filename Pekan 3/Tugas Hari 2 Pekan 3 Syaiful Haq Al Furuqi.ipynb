Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches(9, 6)
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
ax.plot(x, cos, linestyle='solid', color='red')
ax.plot(x, cos3, linestyle='--', color='green')
ax.plot(x, sin, linestyle='dashdot', color='blue')
ax.plot(x, sin2, marker='x', linestyle=' ', color='yellow')
ax.legend(labels=['Cosx','3 cosx','Sin(x-90)','2 Sinx'], loc='center', fancybox=True, framealpha=0.5)

plt.show()
```
